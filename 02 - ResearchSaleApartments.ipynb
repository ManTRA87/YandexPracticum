## Предобработка данных

### Работа с пропущенными значениями

<div class="alert alert-block alert-info">
Обработка пропущенных значений в столбце ceiling_height (высота потолков):
</div>

```python
ceiling_height_median = df['ceiling_height'].median().round(2)
df['ceiling_height'] = df['ceiling_height'].fillna(ceiling_height_median)
```

<div class="alert alert-block alert-info">
Обработка пропущенных значений в столбце floors_total и locality_name (всего этажей):
</div>

```python
df.dropna(subset=['floors_total', 'locality_name'], inplace=True)
```

<div class="alert alert-block alert-info">
Обработка пропущенных значений в столбце living_area (жилая площадь):
    </div>

```python
living_to_total_median = df['living_area'].median()/df['total_area'].median()
df['living_area'] = df['living_area'].fillna(df['total_area'] * living_to_total_median)
```

<div class="alert alert-block alert-info">
Обработка пропущенных значений в столбце is_apartment (апартаменты):
    </div>

```python
df['is_apartment'] = df['is_apartment'].fillna(False)
```

<div class="alert alert-block alert-info">
Обработка пропущенных значений в столбце kitchen_area (площадь кухни):
    </div>

```python
kitchen_to_living_median = df['kitchen_area'].median()/df['living_area'].median()
df['kitchen_area'] = df['kitchen_area'].fillna(df['total_area'] * kitchen_to_living_median)
```

<div class="alert alert-block alert-info">
Обработка пропущенных значений в столбце balcony (число балконов):
    </div>

```python
df['balcony'] = df['balcony'].fillna(0)
```

<div class="alert alert-block alert-info">
Обработка пропущенных значений в столбцах с гегорафической привязкой объектов недвижимости к значимым местам:
    </div>

```python
def fillna_0(item):
    df[item] = df[item].fillna(0)

fillna_0('airports_nearest')
fillna_0('parks_around3000')
fillna_0('parks_nearest')
fillna_0('ponds_around3000')
fillna_0('ponds_nearest')

citi_cityCenters_nearest_NaN = df.query('cityCenters_nearest != cityCenters_nearest')['locality_name'].unique()
#print(citi_cityCenters_nearest_NaN)
#Нахождение самого удаленного объекта
print('Расстояние до центра для самого удаленного объекта недвижимости: ', df['cityCenters_nearest'].max())
citi_non_SPB = df.query('(locality_name != "Санкт-Петербург") & (cityCenters_nearest == "nan")')

df.loc[(df['cityCenters_nearest'].isna())&(df['locality_name'] != "Санкт-Петербург"), 'cityCenters_nearest'] = 70000
fillna_0('cityCenters_nearest')
```

<div class="alert alert-block alert-info">
Обработка пропущенных значений в столбце days_exposition (сколько дней было размещено объявление):
</div>

```python
fillna_0('days_exposition')
```

### Вывод

    Пропуски в столбце ceiling_height заменены медианным значением, так как вероятная причина появления пропусков - отсутсвие информации у пользователей о высоте потолка.
    В данных столбца floors_total есть пропущенные значения, они составляют порядка 0,4% от выборки. Из-за незначительной доли и трудоемкости выявления закономерности, принято решение удалить эти данные из выборки.
    Для заполнения пропущенных значений в столбце living_area принято решение определить сколько в среднем составляет жилая площадь от общей площади. Для этого определены медианные велечины столбцов total_area и living_area, найдено их соотношение, а пропуски заполнены значениями total_area, умноженными на долю жилой площади к общей (в нашем примере доля жилой площади к общей получилась равной 0,58). Аналогично заполнены пропущенные значения для столбца kitchen_area.
    Значения столбца is_apartment логического типа, заполнено только 12% выборки, из них 98% это False. Предположим, что остальные пропущенные значения - это не апартаменты, поэтому заполним их значениием False.
    Пропущенные значения в столбце balcony заменены на 0, так как есть вероятность, что для большей части объектов недвижимости, где данные о количестве балконов не указаны, балконов нет.
    В столбце locality_name пропущенные значения составляют 0,2%. Принято решение их удалить.
    Отсутствующие данные в столбцах о географической привязки объектов недвижимости к определенному месту (аэропорту, ближайшему парку, центру города Санкт-Петербург, прудам) пропущены не пользователем, а автоматизированной системой, заполнявшей данную информацию. Большинство пропущенных данных относятся к загородным населенным пунктам, поэтому, можно предположить, что в автоматизированной системе указан лимит радиуса от значимого места, дальше которого она не заполняет информацию, в связи с чем принято решение заменить пропуски на 0 для всех географических характеристик, кроме расстояния до центра, так как оно повлияет на ход дальнейших исследований. В ходе исследования объектов, географические характеристики которых отсутствуют в выборке, обнаружено, что большинство этих объектов являются Ленинградской областью, поэтому для них пропущенные значения заполнены расстоянием, превышающим максимально известное расстояние от центра по данным выборки - 70 км.
<div class="alert alert-block alert-info">
    В столбце days_exposition пропущенные значения связаны с тем, что на момент формирования выборки, объявления еще не были сняты. Но ввиду малой доли таких прпусков принято решение заменить их на 0.

### Изменение типов данных

<div class="alert alert-block alert-info">
Приведение к uint8 (целые числа в диапазоне от 0 по 255)
    </div>

```python
df['total_images'] = df['total_images'].astype('uint8')
df['rooms'] = df['rooms'].astype('uint8')
df['ceiling_height'] = df['ceiling_height'].astype('uint8')
df['floors_total'] = df['floors_total'].astype('uint8')
df['floor'] = df['floor'].astype('uint8')
df['balcony'] = df['balcony'].astype('uint8')
df['kitchen_area'] = df['kitchen_area'].astype('uint8')
```

<div class="alert alert-block alert-info">
Приведение к uint16 (целые числа в диапазоне от 0 по 65535)
    </div>

```python
df['total_area'] = df['total_area'].astype('uint16')
df['living_area'] = df['living_area'].astype('uint16')
df['parks_nearest'] = df['parks_nearest'].astype('uint16')
df['ponds_nearest'] = df['ponds_nearest'].astype('uint16')
df['days_exposition'] = df['days_exposition'].astype('uint16')
```

<div class="alert alert-block alert-info">
Приведение к uint32 (целые числа в диапазоне от 0 по 4294967295)
    </div>

```python
df['airports_nearest'] = df['airports_nearest'].astype('uint32')
df['cityCenters_nearest'] = df['cityCenters_nearest'].astype('uint32')
```

<div class="alert alert-block alert-info">
Приведение к категориальному типу
    </div>

```python
df['parks_around3000'] = df['parks_around3000'].astype('category')
df['ponds_around3000'] = df['ponds_around3000'].astype('category')
```

<div class="alert alert-block alert-info">
Приведение к логическому типу
    </div>

```python
df['is_apartment'] = df['is_apartment'].astype(bool)
```

<div class="alert alert-block alert-info">
Приведение к временному формату
    </div>

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'])
```
